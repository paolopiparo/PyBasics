```python
# TODO: 1st exercise: Define the `forward_price` function
spot = float(100)
interest_rate = float(0.05)
time = float(3) #in years

#the two functions above are the same
forward_price = spot * e**(interest_rate*time)
forward_price2 = spot * math.exp(interest_rate*time)

print(f"The Forward Price is {round(forward_price2,5)}")
```

```python
# TODO: 2nd exercise: Define the `short_pnl` function

#we will make a profit only if the strike price 
#(at t == 0) is above the spot price at maturity (t == T)

positions = [100,140,200]
mtm = [110,120,180]


total = list(map(operator.sub, positions,mtm))
short_pnl = sum(total)
print(f"The total P&L of our short portfolio is {short_pnl}")
```

```python
# Dataset 
# here we present the number of stocks we bough and at what price 
aapl = [ 10, 154.12 ]
goog = [  2, 812.56 ]
tsla = [ 12, 342.12 ]
fb   = [ 18, 209.0  ]

# portfolio is a nested array. 
# portfolio is a variable containing a list, 
# and each element of that list is a list too!

portfolio = [ aapl, goog, tsla, fb ]
print(portfolio)
```

```python
# How many stocks do I own of Facebook?
print(portfolio[3][0])
```

```python
# Additional Dataset 
# Current mkt stock prices for aapl, goog, tsla and fb 
market = [ 198.84, 1217.93, 267.66, 179.06 ]
```

```python
# Loop functions that add to each portfolio the number of stocks and the prices 
number_of_stocks = []
for iz in range(0, 4):
    number_of_stocks.append(portfolio[iz][0])
    
market_portfolio = []   #Creates empty list for the market portfolio with my number of shares
for i in range(0, len(market)):
     market_portfolio.append(market[i]*number_of_stocks[i])
        
my_portfolio = []   #Create empty list for my portfolio 
for ix in range(0, 4):
     my_portfolio.append(round(portfolio[ix][0]*portfolio[ix][1],5))
```

```python
# Now we multiply in each portfolio the number of stocks * the prices 
# (PTF and Market, i.e. P0 and P1 at sell point)

portfolio_pnl = round(sum(market_portfolio),2) - round(sum(my_portfolio),2)
print(f'My Portfolio was valued {round(sum(my_portfolio),2)} and now is valued {round(sum(market_portfolio),2)}. Therefore my P&L is currently {portfolio_pnl}')
```

```python
#EXAMPLE with metaweather
#/api/location/search/?query=london  --> get the query from here and put it in the second link
#/api/location/44418

import requests

BASE_URL = "https://www.metaweather.com"    #it is constant

city = input("City?\n> ")    #-->  \n is just for the output 
url = f"{BASE_URL}/api/location/search/?query={city}"

#response.content in the Console of Spyder to see the response

response = requests.get(url)   #get HTTP request
print(response.status_code)    #e.g. 404 (famous error code)
print(response.content)        #will be a string with the actual raw data

data = response.json()   #as we know it is a JSON ("data" is a Py list or dict)

#to access firt element in the list I use, accessing the key "woeid"
woeid = data[0]['woeid'] 

url = f"{BASE_URL}/api/location/{woeid}/"
response = requests.get(url)
data = response.json()
print(data)

#I access a List and I want to loop to access the data 
print(data["consolidated_weather"])
#I want trhee informations, weather state name, max temperature and date
for day in data["consolidated_weather"]:
    description = day["weather_state_name"]
    date = day["applicable_date"]
    temp = day["max_temp"]
    print(f"{date}: #{description} (#{round(temp, 2)} in Celsius)")
```

```python
url = "http://books.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

#if we navigate the website, the part cointaining the books is "Article and Class: Product_pod"
books_html = soup.find_all("article", class_="product_pod")
len(books_html)
```

```python
#let's parse one book like with the movies before
books_html[0]
books_html[0].find("h3").find("a")

#now we search the attributes: IMPORTANT the attributes part  
books_html[0].find("h3").find("a").attrs
book_title = books_html[0].find("h3").find("a").attrs["title"]


#to get the price as well 
book_price = books_html[0].find("p", class_="price_color").string
print(f"The Book title is: {book_title} and the book price is: {book_price}")
```

```python
#for the price to get rid of the POUND sign and leave it as a float
book_price = float(books_html[0].find("p", class_="price_color").string[1:])
book_price
```

```python
#then we want the rating (in stars) as well 
#define a function for the stars 
def parse_rating(rating_classes):
    if 'One' in rating_classes:
        return 1
    elif 'Two' in rating_classes:
        return 2
    elif 'Three' in rating_classes:
        return 3
    elif 'Four' in rating_classes:
        return 4
    elif 'Five' in rating_classes:
        return 5
    else:
        return 0
    
book_rating = parse_rating(books_html[0].find("p", class_="star-rating").attrs['class'])
book_rating
```

```python
#LOOP for all books 
books_dict = { 
    'Title': [], 
    'Price': [], 
    'Rating': [] 
}

for book in books_html:
    title = book.find("h3").find("a").attrs["title"]
    price = float(book.find("p", class_="price_color").text[1:])
    rating = parse_rating(book.find("p", class_="star-rating").attrs['class'])
    books_dict["Title"].append(title)
    books_dict["Price"].append(price)
    books_dict["Rating"].append(rating)
```

```python
len(books_dict)   #number of dictionaries 
```

```python
#load data in pandas 
books_df = pd.DataFrame.from_dict(books_dict)
books_df
```

```python
#plot ratings by grouping
books_df.groupby("Rating").count()["Title"].plot(kind="bar")
```

```python
#this was done for the first page only. Loop for more 
page = 1
url = f"http://books.toscrape.com/catalogue/page-{page}.html"
url
    
all_books_dict = { 'Title': [], 'Price': [], 'Rating': [] }

MAX_PAGE = 30
for page in range(1, MAX_PAGE + 1):
    print(f"Parsing page {page}...")
    url = f"http://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    for book in soup.find_all("article", class_="product_pod"):
        title = book.find("h3").find("a").attrs["title"]
        price = float(book.find("p", class_="price_color").text[1:])
        rating = parse_rating(book.find("p", class_="star-rating").attrs["class"])
        all_books_dict["Title"].append(title)
        all_books_dict["Price"].append(price)
        all_books_dict["Rating"].append(rating)

print("Done!")
```

```python
len(all_books_dict["Title"])    #check if we have MAX_PAGE * 20 Books 
```

```python
#load all in a dataframe 
all_books_df = pd.DataFrame.from_dict(all_books_dict)
all_books_df.head()
```

```python
all_books_df["Price"].hist()
```

```python
all_books_df.groupby("Rating").count()["Title"].plot(kind="bar")
```

```python
#Save data into a CSV
all_books_df.to_csv("books.csv")
```
